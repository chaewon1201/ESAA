### 넘파이
- 자연과학, 공학에서 사용
- 빠른 배열 연산 속도를 보장

### 넘파이 ndarray 개요
- import numpy
- 다차원 배열을 쉽게 생성하고 다양한 연산 수행 가능
- Shape 변수 ; 크기 -> 행과 열의 수를 튜플 형대로 가지고 있음 -> 배열의 차원 알 수 있음
(명확한 1차원 데이터 혹은 다차원 데이터를 요구하는 경우  : 명확한 차원의 차수를 변환 방법을 알아야 오류 방지 가능)

#### ndarray 의 데이터 타입
- 숫자 : int형, unsigned형, float형, complex 타입
- 문자열
- 불

: 연산의 특성상 같은 데이터 타입만 가능 / 데이터 타입은 dtype 속성으로 확인 가능

: 리스트를 ndarray로 변경시 더 큰 데이터 타입으로 형변환을 일광적용

- astype() 메서드를 통해 알 수 있음 -> 인자로 원하는 타입을 문자열로 지정하면 됨
- 메모리로 데이터를 전체 로딩한 다음 이를 기반으로 알고리즘 적용 -> 대용량의 데이터 로딩시 수행속도가 느려지거나 메모리 부족으로 오류 발생 가능


### ndarray 편리하게 생성하기 -> arange, zeros, ones
- arange() : range()와 유사 기능 / default 함수 인자는 stop값, 1차원 ndarray를 만들어줌
- zeros() : 튜플 형태의 shape값을 입력 -> 모든 값을 0으로 채운 해당 shape의 ndarray를 생성
- ones()

### ndarray의 차원과 크기 변경 : resahpe()
- 지정 사이즈로 변경 불가시 오류발생 / 호환되는 새오룬 shape으로 변환
- ndarray는 stack or concat으로 결합시 각각의 ndarray의 형태를 통일해 유용하게 사용

### 넘파이의 ndarray의 데이터 세트 선택하기 :  indexing
- 특정 데이터만을 추출 : 원하는 위치의 인덱스 값을 지정하면 해당 위치의 데이터가 반환
- 슬라이싱 : 연속된 인덱스 상의 ndarray를 추출하는 방식
- 팬시 인덱싱 : 해당 위치에 있는 데이터의 ndarray를 반환
- 불린 인덱싱 :  T/F 값 인덱싱 집합을 기반으로 T에 해당하는 인덱스 위치의 ndarray를 반환

##### 단일 값 추출
- 1차원과 2차원 ndarray에서의 데이터 접근의 차이는 2차원의 경우 ,로 분리된 로우와 칼럼 위치의 인덱스를 통해 접근하는 것
- axis 0 ; 로우 방향의 축을 의미 / axis 1은 칼럼 방향의 축 / 3차원의 경우 axis 2도 존재

##### 슬라이싱
- ':' 를 이용
- 1차원 ndarray에 슬라이싱을 적용 -> 유사하게 row, col 각각의 인덱스에 슬라이싱 적용

##### 불린 인덱싱
- 조건 필터링과 검색을 동시에 가능
- 단순히 []내에 원하는 필터링 조건만 넣으면 해당 조건을 만족하는 데이터 세트를 반환

##행렬의 정렬 - sort() & argsort
- 행렬 정렬 : np.sort()와 같이 넘파이에서 sort()를 호출하는 방식과 ndarray.sort()와 같은 행렬 자체에서 sort()를 호출하는 방식
- 행렬이 2차원 이상일 경우 : axis 축 값 설정을 통해 로우방향, 또는 칼럼 방향으로 정렬 수행
 - 정렬 행렬의 인덱스 반환 :  기존 원본 행렬의 원소에 대한 인덱스 필요시 np.argsort()를 이용
 - 넘파이의 ndarray는 메타 데이터를 가질 수 없음 - 실제 값과 그 값이 뜻하는 메타 데이터를 별도의 ndarray로 각각 가져야함

## 선형대수 연산 - 행렬 내적과 전치 행렬 구하기
- 행렬내적 (행렬곱) ; np.dot()
- 두 행렬 A, B의 내적은 왼쪽 행렬의 로우와 오른쪽 행렬의 칼럼의 원소들을 순차적으로 곱한뒤 그 결과를 모두 더한 값
- 전치 행렬 ; 원 행렬에서 행과 열 위치를 교환한 원소로 구성한 행렬 - A^t 로 표기 / transpose()를 이용

```python
org_array = np.array([ 3, 1, 9, 5])
print('원본 행렬:', org_array)
# np.sort()로 정렬
sort_array1 = np.sort(org_array)
print('np.sort() 호출 후 반환된 정렬 행렬:', sort_array1)
print('np.sort() 호출 후 원본 행렬:', org_array)
# ndarray.sort()로 정렬
sort_array2 = org_array.sort()
print('org_array.sort() 호출 후 반환된 행렬:', sort_array2)
print('org_array.sort() 호출 후 원본 행렬:', org_array)
```

```python
sort_array1_desc = np.sort(org_array)[::-1]
print('내림차순으로 정렬:', sort_array1_desc)
```

```python
array2d = np.array([[8, 12],
                    [7, 1]])

sort_array2d_axis0 = np.sort(array2d, axis=0)
print('로우 방향으로 정렬:\n', sort_array2d_axis0)

sort_array2d_axis1 = np.sort(array2d, axis=1)
print('칼럼 방향으로 정렬:\n', sort_array2d_axis1)
```

```python
org_array = np.array([ 3, 1, 9, 5])
sort_indices = np.argsort(org_array)
print(type(sort_indices))
print('행렬 정렬 시 원본 행렬의 인덱스:', sort_indices)
```

```python
org_array = np.array([ 3, 1, 9, 5])
sort_indices_desc = np.argsort(org_array)[::-1]
print('행렬 내림차순 정렬 시 원본 행렬의 인덱스:', sort_indices_desc)
```

```python
import numpy as np

name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array = np.array([78, 95, 84, 98, 88])

sort_indices_asc = np.argsort(score_array)
print('성적 오름차순 정렬 시 score_array의 인덱스:', sort_indices_asc)
print('성적 오름차순으로 name_array의 이름 출력:', name_array[sort_indices_asc])
```

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])
dot_product = np.dot(A, B)
print('행렬 내적 결과:\n', dot_product)
```

```python
A = np.array([[1, 2],
              [3, 4]])
transpose_mat = np.transpose(A)
print('A의 전치 행렬:\n', transpose_mat)
```
