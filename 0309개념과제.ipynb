##판다스
- 행과 열로 구성 (Row x column)
- Data frame (여러개의 행과 열로 이뤄진 2차원 데이터를 담는 데이터 구조체)
- CSV, Tab와 같은 다양한 유형의 분리 문자로 칼럼을 분리한 차이을 손쉽게 Dataframe으로 로딩할 수 있음

sereies : 칼럼이 하나 뿐인 데이터 구조체 / Dataframe은 칼럼이 여러개인 데이터 구조체

- read_csv() ; 어떤 필드 구분 문자 기반의 파일 포맷도 변환 가능
- filepath ; 로드하려는 데이터 파일의 경로를 포함한 파일명을 입력

데이터 파일의 첫번째 줄에 있던 칼럼 문자열이 Dataframe의 칼럼으로 할당

- read_csv() = 파일의 맨 처음 로우를 칼럼명으로 인지하고 칼럼으로 변환

- info() = 메서드를 통해서 총 데이터 건수와 데이터 타입, null 건수를 알 수 있음

- describe() = 개략적인 수준의 분포도를 알 수 있음

- head() = series의 index와 series의 데이터값을 구할 수 있음 (모든 series와 Dataframe을 반드시 가짐)

- value counts : null 값을 포함하여 개별 데이터 값의 건수를 계산할지를 dropna인자로 판단

##Dataframe과 리스트, 딕셔너리, 넘파이 상호변환
- Dataframe (칼럼명을 가지고 있음) -> 리스트, 딕셔너리, ndarray (가능)
- 넘파이로 변환하는 것은 values를 이용
- series 에 상숫값을 할당하면 series의 모든 데이터 세트에 일괄 적용

##Dataframe 데이터 삭제
 - drop()메서드 이용
 axis0 = row / axis1 = column (삭제 시 설정 방법)
 - Dataframe은 유지하고 dataframe을 새롭게 객체 변수로 받고 싶다면 inplace = false로 설정
 - 드롭된 결과 적용시 inplace = True

 ## index 객체
 - series 객체에서 연산 함수를 적용할 때 index는 연산에서 제외 / index는 오직 식별용으로 사용

 ## Data frame의 연산자
 - [] : 행의 위치, 열의 위치, 슬라이싱 범위 등을 지정해 데이터를 가져올 수 있음
 - []내 입력값은 칼럼명을 지정, 칼럼 지정 연산 또는 인덱스 용도로 사용
- iloc[] ; 위치기반 인덱싱은 행과 열 위치를, 0을 출발점으로 하는 세로축, 가로축 좌표 정숫값으로 지정하는 방식
- loc[] : 명칭 기반으로 데이터를 추출

```python
#나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young Adult'
  elif age <= 60: cat = 'Adult'
  else: cat = 'Elderly'

  return cat

#lambda 식에 위에서 생성한 get_category()함수를 반환값으로 지정
#get_category(X)는 입력값으로 'Age'칼럼 값을 받아서 해당하는 cat반환

titanic_df['Age_Cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age', 'Age_Cat']].head()
```
